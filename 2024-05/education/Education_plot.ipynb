```python
import matplotlib.pyplot as plt
import numpy as np

# Data
constructs = ["Confidence", "Usefulness", "Interest", "Professional Perception"]
control_pre = [3.2, 3.5, 3.01, 2.8]
control_post = [3.3, 3.6, 3.0, 2.9]
experimental_pre = [3.1, 3.3, 3.2, 2.7]
experimental_post = [3.8, 4.2, 4.5, 3.2]

# Standard deviations (error bars)
std_control_pre = [0.5, 0.6, 0.4, 0.5]
std_control_post = [0.4, 0.5, 0.3, 0.4]
std_experimental_pre = [0.6, 0.5, 0.4, 0.6]
std_experimental_post = [0.5, 0.3, 0.2, 0.5]

# Bar width
bar_width = 0.2

# X-axis positions
x = np.arange(len(constructs))

# Creating the bar plot
fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - 1.5 * bar_width, control_pre, bar_width, label="Control Pre", color='#FF6666', yerr=std_control_pre, capsize=5)
bars2 = ax.bar(x - 0.5 * bar_width, control_post, bar_width, label="Control Post", color='#990000', yerr=std_control_post, capsize=5)
bars3 = ax.bar(x + 0.5 * bar_width, experimental_pre, bar_width, label="Experimental Pre", color='blue', yerr=std_experimental_pre, capsize=5)
bars4 = ax.bar(x + 1.5 * bar_width, experimental_post, bar_width, label="Experimental Post", color='darkblue', yerr=std_experimental_post, capsize=5)

# Labels and title with increased font size
ax.set_xlabel("", fontsize=16, labelpad=15)  # Increase font size and add padding
ax.set_ylabel("Mean Likert Score", fontsize=16, labelpad=15)  # Increase font size and add padding
ax.set_xticks(x)
ax.set_xticklabels(constructs, fontsize=12)  # Increase x-tick label font size
ax.set_ylim(2.5, 5)
ax.yaxis.set_tick_params(labelsize=16)  # Increase y-tick label font size

# Move the legend inside the plot (upper right corner) and increase font size
ax.legend(loc='upper right', frameon=True, facecolor='white', edgecolor='black', fontsize=12)

# Apply grid settings similar to Plotly
ax.grid(True, which='major', linestyle=':', linewidth=1, color='lightgrey')

# Show the plot
plt.tight_layout()
plt.savefig('education.jpeg')
plt.show()
```
